```python
%gui tk

import pandas as pd
import nltk
import tkinter as tk
from tkinter import messagebox
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

nltk.download('punkt')
nltk.download('stopwords')

# ==============================
# Dataset
# ==============================

positive_reviews = [
"The food was absolutely amazing and full of flavor",
"I really enjoyed the delicious biryani served here",
"Excellent service and tasty food",
"The ambience was beautiful and relaxing",
"Food quality was top notch"
]

negative_reviews = [
"The food was very bad and tasteless",
"Service was extremely slow",
"The staff was rude and unprofessional",
"Food was served cold",
"Worst dining experience ever"
]

reviews = positive_reviews + negative_reviews
labels = ["Positive"] * len(positive_reviews) + ["Negative"] * len(negative_reviews)

df = pd.DataFrame({"Review": reviews, "Sentiment": labels})

# ==============================
# Preprocessing
# ==============================

stop_words = set(stopwords.words('english'))

def preprocess(text):
    text = text.lower()
    tokens = nltk.word_tokenize(text)
    tokens = [w for w in tokens if w.isalpha() and w not in stop_words]
    return " ".join(tokens)

df["Cleaned"] = df["Review"].apply(preprocess)

# ==============================
# Train Model
# ==============================

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["Cleaned"])
y = df["Sentiment"]

model = LogisticRegression(max_iter=3000)
model.fit(X, y)

# ==============================
# Global Variable
# ==============================

latest_result = "No Review Yet"

# ==============================
# Customer Window
# ==============================

def open_customer():
    customer_window = tk.Toplevel(root)
    customer_window.title("Customer Panel")
    customer_window.geometry("400x250")

    tk.Label(customer_window, text="Enter Your Review",
             font=("Arial", 14)).pack(pady=10)

    review_entry = tk.Entry(customer_window, width=40)
    review_entry.pack(pady=10)

    def submit():
        global latest_result
        review = review_entry.get()

        if review == "":
            messagebox.showwarning("Warning", "Please enter review!")
            return

        clean = preprocess(review)
        vector = vectorizer.transform([clean])
        prediction = model.predict(vector)[0]

        latest_result = prediction
        messagebox.showinfo("Success", "Review Submitted!")

    tk.Button(customer_window, text="Submit",
              command=submit).pack(pady=10)

# ==============================
# Owner Window
# ==============================

def open_owner():
    owner_window = tk.Toplevel(root)
    owner_window.title("Owner Panel")
    owner_window.geometry("400x250")

    tk.Label(owner_window, text="Latest Customer Sentiment",
             font=("Arial", 14)).pack(pady=20)

    color = "green" if latest_result == "Positive" else "red"

    tk.Label(owner_window, text=latest_result,
             font=("Arial", 18, "bold"),
             fg=color).pack(pady=20)

# ==============================
# Main Window
# ==============================

root = tk.Tk()
root.title("Restaurant Review System")
root.geometry("400x300")

tk.Label(root, text="Select User Type",
         font=("Arial", 16, "bold")).pack(pady=20)

tk.Button(root, text="Customer Panel",
          width=20, height=2,
          command=open_customer).pack(pady=10)

tk.Button(root, text="Owner Panel",
          width=20, height=2,
          command=open_owner).pack(pady=10)

root.mainloop()
```
